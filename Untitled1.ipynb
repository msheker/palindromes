```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_file
from bokeh.layouts import gridplot, Row
from bokeh.io import hplot
```

```python
data = pd.read_excel("RAW_DATA-2iwcznn-2kr2xw0.xls")

del data["BMI"]
del data["cholesterol"]
del data["high_d_lipid"]
del data["hba1c"]
del data["mean_syst"]
del data["mean_diast"]
del data["bmi"]
del data["low_d_lipid"]
del data["R22_bpgroup"]

# remove 41 unknown hiv results
data = data[data["hiv"] != "unknown"]

ls = []
for i in data["hiv"]:
    if i == "negative":
        ls.append(0)
    else:
        ls.append(1)

data["hiv"] = ls

ls = []
for i in data["age_yrs"]:
    if i == "80+":
        ls.append(80)
    else:
        ls.append(float(i))

data["age_yrs"] = ls    
    
data
```

```python
data_ = data[data["age_yrs"] > 19]
data_ = data_[data_["age_yrs"] < 61]
data_ = data_[data_["cmv"] > .3]

data_0 = data_[data_["hiv"] == 0]
data_1 = data_[data_["hiv"] == 1]

p = figure(title = "CMV vs Age")
p.yaxis.axis_label = 'CMV'
p.xaxis.axis_label = 'Age'

p.circle(data_1["age_yrs"], data_1["cmv"], color="red", size=4)
p.circle(data_0["age_yrs"], data_0["cmv"], color="blue", size=2)

p.line(x = [ x for x in range(20,61)], y = [np.mean(data_0["cmv"]) for x in range(81)], color = "blue")
p.line(x = [ x for x in range(20,61)], y = [np.mean(data_1["cmv"]) for x in range(81)], color = "red")

show(p)
```

```python
data0 = data[data["hiv"] == 0]
data1 = data[data["hiv"] == 1]

p1 = figure(title = "HIV Negative")
p1.yaxis.axis_label = 'CMV'
p1.xaxis.axis_label = 'Age'

p1.circle(data0["age_yrs"], data0["cmv"], size=2)
p1.line(x = [ x for x in range(81)], y = [1.2161 for x in range(81)], color = "red")
p1.line(x = [ x for x in range(81)], y = [.9231 for x in range(81)], color = "red")
p1.line(x = [ x for x in range(81)], y = [0.3 for x in range(81)], color = "red")

p2 = figure(title = "HIV Postive")
p2.yaxis.axis_label = 'CMV'
p2.xaxis.axis_label = 'Age'

p2.circle(data1["age_yrs"], data1["cmv"], size=2)
p2.line(x = [ x for x in range(81)], y = [1.216 for x in range(81)], color = "red")
p2.line(x = [ x for x in range(81)], y = [.9231 for x in range(81)], color = "red")
p2.line(x = [ x for x in range(81)], y = [0.3 for x in range(81)], color = "red")

show(gridplot(p1, p2, ncols=2, plot_width=400, plot_height=400, toolbar_location=None))
```
